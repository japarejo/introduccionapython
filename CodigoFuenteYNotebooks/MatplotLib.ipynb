# Matplotlib

Matplotlib es una biblioteca de gráficos de bajo nivel en python que sirve como utilidad de visualización.

Matplotlib fue creada por John D. Hunter.

Matplotlib es de código abierto y podemos utilizarla libremente.

Matplotlib está escrito principalmente en python, algunos segmentos están escritos en C, Objective-C y Javascript para alcanzar la máxima eficiencia y compatibilidad con otras plataformas.

Matplotlib es un paquete no estándar, por lo que deberemos instalarlo con pip si queremos usarlo (a no ser que usemos Anaconda, que nos proporciona una versión pre-instalada).

La mayoria de las funciones de Matplotlib se basan en el módulo pyplot, que normalmente se importa con el alias plt:

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

La función plot() se utiliza para dibujar puntos (marcadores) en un diagrama.

Por defecto, la función plot() dibuja una línea de punto a punto.

La función toma parámetros para especificar los puntos en el diagrama.

El parámetro 1 es una matriz que contiene las coordenadas de los puntos en el eje x.

El parámetro 2 es una matriz que contiene las coordenadas de los puntos del eje Y.

Si necesitamos trazar una línea desde (1, 3) hasta (8, 10), tenemos que pasar dos listas [1, 8] y [3, 10] a la función de trazado.

Para pintar únicamente los puntos, sin una línea que los una, podemos usar un tercer parámetro de la función *plot*:

```python
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

Podemos pintar tantos puntos como deseemos, la única restricción es que las listas de coordenadas que proporcionemos tengan las mismas dimensiones:

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

Si no proporcionamos las coordenadas de las x, se colocarán por defectos en las posiciones correspondientes del array, es decir, 0,1,2,3,4,...

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

Podemos usara el agrumento marker para enfatizar la localización de los puntos:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

También se puede utilizar el parámetro de notación de cadena abreviada para especificar el marcador.

Veamos un ejemplo:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

| Line Syntax |     Description    |

| '-'         | Solid line         |

Además podemos usar el parámetro *markersize*, o su sinónimo *ms*, para controlar el tamaño de los marcadores:

```python
points = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

También tenemos disponible el parámetro *makeredgecolor*, o su sinónimo *mec* para establecer el color del borde de los marcadores:

```python

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

El relleno de los marcadores se controla con el parámetro *markerfacecolor* o *mfc*:

```python
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

Podemos usar las coordenadas RBG en hexadecimal del color (<a href="https://www.w3schools.com/colors/colors_hexadecimal.asp">aquí</a> tienes un calculador para escoger el color que quieras en ese formato) para especificar colores no estándar:

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

Aparte de los colores básicos en formato corto arriba listados, tenemos disponibles otros <a href="https://www.w3schools.com/colors/colors_names.asp">140 colores adicionales</a>:

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
```

Otro parámetro interesante es *linewidth* o *lw* que controla el grosor de la línea en puntos:

```python
plt.plot(ypoints, linewidth = '20.5')
```

Podemos pintar tantas líneas como deseemos en el mismo gráfico simplemente haciendo otra llamada al método *plot*:

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

Podemos añadir etiquetas a los ejes x e y, e incluo un título para nuestro gráfico:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Datos deportivos")
plt.xlabel("Pulso medio")
plt.ylabel("Calorías quemadas por unidad de tiempo")

plt.show()
```

Podemos incluso usar el parámetro fontdict de la funciones anteriores para controlar la propiedades del título y las etiquetas

```python

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

El parámetro *loc* de la función *title* permite controlar la alineación del título. Los valores posibles son 'left', 'right' y 'center'. El valor por defecto es 'center'.

```python
plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

Además de todo lo anterior, para mejorar la precisión con la que el gráfico representa los datos, podemos añadir lineas de cuadrícula al mismo con la función *grid*.

```python
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show() 
```

Esta cuadrícula puede mostrar solo en uno de los ejes si lo deseamos:

```python
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show() 
```

Al igual que el resto de elementos del gráfico, podemos controlar el estilo de las líneas de la cuadrícula con diferentes parámetros:

```python

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

# Gráficos de puntos y mapas de color

Si lo que deseamos mostrar es un gráfico de puntos, podemos usar la función *scatter* en lugar de la función *plot*:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show() 
```

Podemos incluso controlar el color de cada punto individual:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

El módulo Matplotlib tiene una serie de mapas de colores disponibles.

Un mapa de colores es como una lista de colores, donde cada color tiene un valor que va de 0 a 100.

Aquí hay un ejemplo de un mapa de colores:

<img src="https://www.w3schools.com/python/img_colorbar.png"/>

Este mapa de colores se llama 'viridis' y como puedes ver va desde 0, que es un color púrpura, y hasta 100, que es un color amarillo.

Podemos representar una tercera dimensión en el scatterplot especificando los valores asociados al color y el mapa de color a usar:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
z = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=z, cmap='viridis')

plt.show() 
```

Podemos incluir la leyenda del mapa de colores en el dibujo incluyendo la sentencia *plt.colorbar()*:

```python
plt.scatter(x, y, c=z, cmap='viridis')

plt.colorbar()

plt.show() 
```

Existen una gran multitud de mapas de colores predefinidos tal y como se muestra <a href="https://www.analyticsvidhya.com/blog/2020/09/colormaps-matplotlib/">aquí</a>. Además en ese mismo recurso se muestra cómo construir nuestros propios colormaps.

Podemos llegar a presentar hasta dos dimensiones más de cada punto en nuestro scatterplot haciendo uso de dos parámetros adicionales: el tamaño de los puntos y su transparencia. Por ejemplo para mostrar los puntos con distintos tamaños podemos hacer uso del parametro s (de size):

```python

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show() 
```

```python
plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show() 
```

Si combinamos todos los parámetros podemos llegar a general gráficos muy ricos:

```python

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show() 
```

## Diagramas de Barras e Histrogramas

Además de gráficos de puntos, también podemos pitar gráficos de barras:

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

Que también pueden ir en horizontal


```python
plt.barh(x, y)
plt.show()
```

La anchura de las barras es controlable con el parámetro witdh:

```python
plt.bar(x, y, width = 0.1)  # El valor por defecto es 0.8 (medido en términos del espacio relativo entre marcadores).
plt.show()
```

En el caso de que queramos generar histogramas, la libería proporciona una función específica para ello:

```python
import numpy as np

x = np.random.normal(170, 10, 250)

print(x) 

plt.hist(x)
plt.show() 
```

## Diagramas de Sectores

Otro gráfico muy clásico (aunque no se recomienda su uso para casos en que las diferencias de valores sean sutiles) son los diagramas de sectores:

```python

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

¿Quizás quieres que una de las cuñas destaque? El parámetro explode te permite hacerlo.

El parámetro explode, si se especifica, debe ser un array con un valor para cada cuña.

Cada valor representa a qué distancia del centro con la que se muestra cada cuña:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

Podemos añadir una leyenda al gráfico usando la función *legend*:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:",loc="upper left")
plt.show() 
```

Las posiciones posibles para el parámetro loc son: 
- upper right
- upper left
- lower left
- lower right
- right
- center left
- center right
- lower center
- upper center
- center
